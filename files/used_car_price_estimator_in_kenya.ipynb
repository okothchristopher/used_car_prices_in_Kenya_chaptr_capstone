# scrapping multiple pages 

#### controlling the  crawl rate why is this important 
>*Controlling the crawl rate is beneficial for the scraper and for the website we’re scraping. If we avoid hammering the server with a lot of requests all at once, then we’re much less likely to get our IP address banned — and we also avoid disrupting the activity of the website we scrape by allowing the server to respond to other user requests as well*


```python
### we need to be changing the url parameters 
pages=np.arange(1,451,1)# 
pages[2:19]
```

```python
# just to see how the pages would look like 
for page in pages[1:10]:
    page="https://www.cheki.co.ke/cars?page=" + str(page)
    print(page)
```

```python
names=[]
prices=[]
ratings=[]
car_details=[]
for page in pages: 
    page=requests.get("https://www.cheki.co.ke/cars?page=" + str(page))
    soup_cheki=bs(page.text,'html.parser')
    cars=soup_cheki.find_all(class_="right-card-part-container")
    sleep(randint(2,10))# this is to delay the eexecution for a certain number of random seconds 
    for item in cars:
        car_name=item.find(class_='card-header')
        names.append(car_name.text)
        price=item.find(class_='listing-price')
        prices.append(price.text)
        details=item.find(class_='card-features')
        car_details.append(details.text)
```

```python
cheki=pd.DataFrame({'make':names,
                   'price':prices,
                   'details':car_details})
cheki.tail()
```

```python
# some little cleaning
cheki['make']=cheki['make'].map(lambda x: x.replace("\n"," "))
cheki['details']=cheki['details'].map(lambda x: x.replace("\n"," "))
cheki['price']=cheki['price'].map(lambda x: x.replace("\n"," "))
cheki.head()
```
